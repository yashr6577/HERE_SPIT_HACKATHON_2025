```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KDTree
from shapely.geometry import LineString, Point
import ast
import warnings
warnings.filterwarnings('ignore')


def parse_coordinates(coord_str: str):
    """Parse coordinate string into list of tuples"""
    try:
        if pd.isna(coord_str) or coord_str.strip() == '':
            return []
        coords = ast.literal_eval(coord_str)
        if isinstance(coords, tuple) and len(coords) > 0:
            # Single point (POI)
            if isinstance(coords[0], (int, float)):
                return [coords]
            # Multiple points (street linestring)
            else:
                return list(coords)
        return coords
    except Exception:
        return []


def parse_poi_coordinates(lat_val, lon_val):
    """Parse POI coordinates from separate lat/lon columns"""
    try:
        # Handle case where values might be strings or already numeric
        if pd.isna(lat_val) or pd.isna(lon_val):
            return []
        
        # Convert to float if they're strings
        if isinstance(lat_val, str):
            lat = float(lat_val)
        else:
            lat = lat_val
            
        if isinstance(lon_val, str):
            lon = float(lon_val)
        else:
            lon = lon_val
            
        # Return as list containing single coordinate tuple
        return [(lat, lon)]
    except (ValueError, TypeError):
        return []


def linestring_to_points(linestring_coords: list) -> list:
    """
    Convert LINESTRING coordinates to representative points for KD-tree indexing.
    We sample start, end, midpoint, and other key points for better spatial coverage.
    """
    if not linestring_coords:
        return []
    
    points = []
    # Always add start and end points
    points.append(linestring_coords[0])
    points.append(linestring_coords[-1])
    
    n = len(linestring_coords)
    if n > 2:
        # Add midpoint
        points.append(linestring_coords[n // 2])
    if n > 4:
        # Add quarter and three-quarter points
        points.append(linestring_coords[n // 4])
        points.append(linestring_coords[(3 * n) // 4])
    
    # Remove duplicates
    unique_points = list(set(points))
    return unique_points


def calculate_distance_to_linestring(poi_point: tuple, linestring_coords: list) -> float:
    """
    Calculate the shortest distance from a POI point to a LINESTRING in meters.
    Uses a conservative conversion factor for Singapore latitude (~1.3N).
    """
    try:
        poi = Point(poi_point)
        line = LineString(linestring_coords)
        distance_degrees = poi.distance(line)
        # Approximate conversion degrees -> meters:
        # 1 degree latitude  111,000 m
        # 1 degree longitude  110,900 m 
        distance_meters = distance_degrees * 111000   #exact value is 111,139
        return distance_meters
    except Exception:
        return float('inf')

def calculate_street_buffer_width_trial2(street_row):
    """
    Calculate appropriate buffer width based on street characteristics
    
    Parameters:
    - street_row: Row from streets dataframe with lane and divider info
    
    Returns:
    - buffer_width: Width in meters to consider as "on street"
    """
    base_lane_width = 3.2  # meters per lane in Singapore
    
    # Get number of lanes
    lane_cat = str(street_row.get('LANE_CAT', '2'))
    if lane_cat == '1':
        num_lanes = 1
    elif lane_cat == '2':
        num_lanes = 2
    elif lane_cat == '3':
        num_lanes = 3
    else:
        num_lanes = 2  # default
    
    # Base road width
    road_width = num_lanes * base_lane_width
    
    # Check for divider
    divider = str(street_row.get('DIVIDER', 'N')).upper()
    dividerleg = str(street_row.get('DIVIDERLEG', 'N')).upper()
    
    # Add divider width if present
    if divider == 'Y' or dividerleg == 'Y' or divider in ['1', '2', 'L', 'A']:
        road_width += 2.0  # Add 2m for divider width
    
    # Functional class adjustment (highway vs local road)
    func_class = str(street_row.get('FUNC_CLASS', '4'))
    if func_class in ['1', '2']:  # Major roads/highways
        road_width *= 1.2  # 20% wider for major roads
    
    # Buffer is half the road width (from centerline to edge)
    buffer_width = road_width / 2
    
    return buffer_width, road_width, num_lanes


def analyze_poi_placement(distance_meters: float, road_width_m: float = 3.2):
    """
    Classify POI placement relative to the street centerline.
    
    Parameters:
    - distance_meters: Distance to street centerline
    - road_width_m: Average road width in meters
    
    Returns:
    - (placement_status, description)
    """
    if distance_meters is None or distance_meters == float('inf'):
        return 'UNKNOWN', 'No nearest street found'
    
    half_road = road_width_m / 2  # 1.6 meters for half-width
    
    # ON_STREET: within half road width from centerline
    # NEAR_STREET: within sidewalk buffer (~5 m)
    # FAR_FROM_STREET: within reasonable distance (e.g., 50 m)
    # VERY_FAR: beyond reasonable distance
    
    sidewalk_buffer = 5.0
    reasonable_distance = 50.0
    
    if distance_meters <= half_road:
        return 'ON_STREET', f'POI is on the street (within {half_road:.1f} m of centerline)'
    elif distance_meters <= sidewalk_buffer:
        return 'NEAR_STREET', f'POI is near street (sidewalk/building adjacent) at {distance_meters:.1f} m'
    elif distance_meters <= reasonable_distance:
        return 'FAR_FROM_STREET', f'POI is far from street but within reasonable distance ({distance_meters:.1f} m)'
    else:
        return 'VERY_FAR', f'POI is very far from nearest street ({distance_meters:.1f} m) - possible data error'


def find_nearest_linestring(
    streets_df: pd.DataFrame,
    pois_df: pd.DataFrame,
    k_candidates: int = 10,
    road_width_m: float = 3.2
) -> pd.DataFrame:
    """
    For each POI, find nearest street LINESTRING and calculate distance and placement.
    
    Returns DataFrame with additional columns:
    nearest_street_id, nearest_street_name, distance_to_street_meters,
    placement_status, placement_description, is_on_street, is_correctly_placed
    """
    
    print("Building representative points for streets...")
    
    street_points = []
    street_indices = []
    
    for idx, row in streets_df.iterrows():
        coords = parse_coordinates(row['extracted_coords'])
        if coords:
            rep_points = linestring_to_points(coords)
            for pt in rep_points:
                street_points.append(pt)
                street_indices.append(idx)
    
    if not street_points:
        raise ValueError("No valid street coordinates found")
    
    street_points_arr = np.array(street_points)
    kd_tree = KDTree(street_points_arr)
    
    print(f"KD-tree built with {len(street_points)} points from {len(streets_df)} streets")
    
    results = []
    valid_pois = 0
    
    for idx, poi_row in pois_df.iterrows():
        # Parse POI coordinates from separate lat/lon columns
        poi_coords = parse_poi_coordinates(poi_row["DISPLAY_LO"], poi_row["DISPLAY_LA"])
        
        if not poi_coords:
            # Invalid or missing POI coordinates
            result = poi_row.to_dict()
            result.update({
                'nearest_street_id': None,
                'nearest_street_name': None,
                'distance_to_street_meters': None,
                'nearest_link_id': None,
                'placement_status': 'NO_COORDS',
                'placement_description': 'POI has no valid coordinates',
                'is_on_street': False,
                'is_correctly_placed': False
            })
            results.append(result)
            continue
        
        poi_point = poi_coords[0]  # POIs are single points
        valid_pois += 1
        
        # Query KD-tree for nearest street candidate points
        query_k = min(k_candidates, len(street_points))
        distances, indices = kd_tree.query([poi_point], k=query_k)
        
        # Get unique street indices from candidate points
        candidate_street_ids = set(street_indices[i] for i in indices[0])
        
        min_distance = float('inf')
        nearest_street_idx = None
        
        # Calculate exact distances to candidate streets
        for street_id in candidate_street_ids:
            street_coords = parse_coordinates(streets_df.iloc[street_id]['extracted_coords'])
            if street_coords:
                dist = calculate_distance_to_linestring(poi_point, street_coords)
                if dist < min_distance:
                    min_distance = dist
                    nearest_street_idx = street_id
        
        placement_status, placement_desc = analyze_poi_placement(min_distance, road_width_m)
        
        result = poi_row.to_dict()
        if nearest_street_idx is not None:
            nearest_street = streets_df.iloc[nearest_street_idx]
            result.update({
                'nearest_street_id': nearest_street_idx,
                'nearest_street_name': nearest_street.get('ST_NAME', 'Unknown'),
                'distance_to_street_meters': min_distance,
                'nearest_link_id': nearest_street.get('LINK_ID', None),
                'placement_status': placement_status,
                'placement_description': placement_desc,
                'is_on_street': placement_status == 'ON_STREET',
                'is_correctly_placed': placement_status in ['ON_STREET', 'NEAR_STREET']
            })
        else:
            # No nearest street found
            result.update({
                'nearest_street_id': None,
                'nearest_street_name': None,
                'distance_to_street_meters': None,
                'nearest_link_id': None,
                'placement_status': 'UNKNOWN',
                'placement_description': 'No nearest street found',
                'is_on_street': False,
                'is_correctly_placed': False
            })
        
        results.append(result)
        
        # Progress log every 1000 POIs
        if (idx + 1) % 1000 == 0:
            print(f"Processed {idx + 1} POIs... ({valid_pois} with valid coordinates)")
    
    print(f"Total valid POIs processed: {valid_pois}/{len(pois_df)}")
    return pd.DataFrame(results)


def load_and_process_data(streets_file: str, pois_file: str):
    """Load CSV data and find nearest streets for POIs"""
    
    print("Loading street and POI data...")
    streets_df = pd.read_csv(streets_file)
    pois_df = pd.read_csv(pois_file)
    
    print(f"Loaded {len(streets_df)} streets and {len(pois_df)} POIs")
    
    # Debug: Check POI coordinate columns
    print(f"POI columns: {list(pois_df.columns)}")
    if 'DISPLAY_LA' in pois_df.columns and 'DISPLAY_LO' in pois_df.columns:
        print(f"Sample POI coordinates:")
        sample_coords = pois_df[['DISPLAY_LA', 'DISPLAY_LO']].head()
        for idx, row in sample_coords.iterrows():
            print(f"  Row {idx}: lat={row['DISPLAY_LA']}, lon={row['DISPLAY_LO']}")
    else:
        print("Warning: DISPLAY_LA and DISPLAY_LO columns not found!")
        print("Available columns:", list(pois_df.columns))
    
    results_df = find_nearest_linestring(streets_df, pois_df)
    
    return results_df
```

```python
def POI_street_analysis(streets_file, pois_file, output_file):
    try:
        results = load_and_process_data(streets_file, pois_file)
        
        print("\n=== Results Summary ===")
        print(f"Total POIs processed: {len(results)}")
        print(f"POIs with nearest street found: {results['nearest_street_id'].notna().sum()}")
        
        valid_distances = results['distance_to_street_meters'].dropna()
        if not valid_distances.empty:
            print(f"Average distance to nearest street: {valid_distances.mean():.2f} meters")
            print(f"Median distance to nearest street: {valid_distances.median():.2f} meters")
            print(f"Max distance to nearest street: {valid_distances.max():.2f} meters")
        
        print("\n=== Placement Analysis ===")
        placement_counts = results['placement_status'].value_counts()
        for status, count in placement_counts.items():
            pct = count / len(results) * 100
            print(f"{status}: {count} POIs ({pct:.1f}%)")
        
        correctly_placed = results['is_correctly_placed'].sum()
        on_street = results['is_on_street'].sum()
        print(f"\nCorrectly placed POIs (on/near street): {correctly_placed} ({correctly_placed / len(results) * 100:.1f}%)")
        print(f"POIs directly on street: {on_street} ({on_street / len(results) * 100:.1f}%)")
        
        print("\n=== Sample Results ===")
        sample_results = results[results['nearest_street_id'].notna()].head(10)
        for _, row in sample_results.iterrows():
            print(f"POI {row.get('OBJECTID', 'N/A')}: {row['placement_status']} - "
                  f"Street '{row['nearest_street_name']}' at {row['distance_to_street_meters']:.1f} m")
            print(f"   {row['placement_description']}")
        
        
        # output_file = "poi_nearest_streets_99.csv"
        # if 'is_correctly_placed' in results.columns:
        #     results = results.drop(columns=['is_correctly_placed'])

        results.to_csv(output_file, index=False)
        print(f"\nResults saved to {output_file}")
        
    except Exception as e:
        print(f"Error: {e}")
        import traceback
        traceback.print_exc()
        print("Please ensure your CSV files exist and have the correct columns.")
```
